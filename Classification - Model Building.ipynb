```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier

from imblearn.combine import SMOTEENN
```

```python
dfTrain = pd.read_csv('TrainData.csv')
dfTrain.head()
```

```python
dfTest = pd.read_csv('TestData.csv')
dfTest.head()
```

```python
dfTrain.drop(['Unnamed: 0'], axis = 1, inplace = True)
dfTest.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
x_train = dfTrain.drop('Exited', axis = 1)
x_train
```

```python
y_train = dfTrain['Exited']
y_train
```

```python
x_test = dfTest.drop('Exited', axis = 1)
y_test = dfTest['Exited']
```

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test, y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
sm = SMOTEENN()
X_trainNew, y_trainNew = sm.fit_sample(x_train,y_train)
X_testNew, y_testNew = sm.fit_sample(x_test,y_test)
```

```python
model_dt_smote=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt_smote.fit(X_trainNew, y_trainNew)
```

```python
yr_predict = model_dt_smote.predict(X_testNew)
model_score_r = model_dt_smote.score(X_testNew, y_testNew)
print(model_score_r)
print(metrics.classification_report(y_testNew, yr_predict))
```
